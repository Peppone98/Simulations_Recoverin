```python
import MDAnalysis as mda
import numpy as np
```

```python

def filter_atoms(input_gro, output_gro, x_range, y_range):
    """
    Read a GROMACS .gro file and create another .gro file containing
    only the atoms within specified distances over the (x, y) coordinates.
    """

    # define the list of True and False for the residues
    res_sel = []

    # Load the .gro file
    u = mda.Universe(input_gro)

    # Iterate over residues
    for residue in u.residues:
    
        # Iterate over atoms within the current residue
        count = 0
        for atom in residue.atoms:

            # The z-axis has no problem. The y=0 is already OK
            x_sel = bool((u.atoms.positions[atom.index, 0] >= x_range[0]) & (u.atoms.positions[atom.index, 0] <= x_range[1]))
            y_sel = bool(u.atoms.positions[atom.index, 1] <= y_range[1])

            # Combine selections to get the final selection
            sel = bool(x_sel & y_sel)
            
            # count the True values:
            if sel == True:
                count = count + 1
   

        if count >= 0.3 * len(residue.atoms[:]):

            # If the atoms appear inside the box with a percentage bigger than 30%, then append a True
            res_sel.append(True)
            print(f"Residue {residue.resname} ({residue.resid}): inside the new box")

        else: 
            res_sel.append(False)
            print(f"Residue {residue.resname} ({residue.resid}): outside the new box")

    # Outside the loop over the residues
    u.atoms[res_sel].write(output_gro)
            
```
