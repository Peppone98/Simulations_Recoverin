```python
import numpy as np
import matplotlib.pyplot as plt
import plumed
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import math
```

```python
plt.rcParams["figure.figsize"] = (5, 3)

DIST_nmRec = plumed.read_as_pandas("DIST_nmRec")
plt.hist(DIST_nmRec["d1"][:1000], bins=80, density=True, alpha = 0.5, color= 'b', label="nmRec")
plt.xlabel(r"Distance Oxygens-Ca$^{2+}$ [nm]")
plt.ylabel("Area normalized bin counts")
plt.title(r"Distribution of $d$ in nmRec")
plt.grid()
plt.minorticks_on()
plt.savefig("../pics/d_nmRec_unb")
```

```python
DIST_nmRec_GRK1 = plumed.read_as_pandas("DIST_nmRec_GRK1")
plt.hist(DIST_nmRec_GRK1["d1"][:1000], bins=80, density=True, alpha = 0.5, color= 'r', label="nmRec with GRK1")
plt.xlabel(r"Distance Oxygens-Ca$^{2+}$ [nm]")
plt.ylabel("Area normalized bin counts")
plt.title(r"Distribution of $d$ in nmRec with GRK1")
plt.grid()
plt.minorticks_on()
plt.savefig("../pics/d_nmRec_GRK1_unb")
```
