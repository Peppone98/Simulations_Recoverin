```python
import numpy as np
import matplotlib.pyplot as plt
import plumed
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import math
```

```python
# Import data from the colvar files
data_norestr = plumed.read_as_pandas("COLVAR_norestr")
data_restr = plumed.read_as_pandas("COLVAR_restr")
```

```python
plt.rcParams["figure.figsize"] = (5, 3)
plt.plot(data_norestr["time"]*0.001, data_norestr["rmsd"], label="no rmsd restraint")
plt.plot(data_restr["time"]*0.001, data_restr["rmsd"], label="restraint")
plt.grid()
plt.title("Metadynamics")
plt.ylabel("rmsd [nm]")
plt.xlabel("Time [ns]")
plt.legend(loc="best")
```

```python
plt.plot(data_norestr["time"]*0.001, data_norestr["uwall-rmsd.bias"], label="no rmsd restraint")
plt.plot(data_restr["time"]*0.001, data_restr["uwall-rmsd.bias"], label="restraint")
plt.grid()
plt.title("Metadynamics")
plt.ylabel("rmsd [nm]")
plt.xlabel("Time [ns]")
plt.legend(loc="best")
```
